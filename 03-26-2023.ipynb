```python
from sympy import Symbol, residue, pi, cot
```

```python
z = Symbol('z', complex = True)
residue(pi * cot(pi * z) / (z * z), z, 0)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0.0, 2.0 * np.pi, 24)
ReG = 1.5 * np.cos(t)
ImG = 1.5 * np.sin(t)
```

```python
fig, ax = plt.subplots(figsize = (3.5, 3.5), dpi = 256)

ax.set_xlim(left = -2.5, right = 2.5)
ax.set_ylim(bottom = -2.5, top = 2.5)

ax.quiver(ReG[:-1], ImG[:-1],
          np.diff(ReG),
          np.diff(ImG), color = 'C0')

for i in range(-2,3):
    ax.scatter(x = i, y = 0, color = 'C1')

fig.savefig('hr9Phs6LzH.jpeg', transparent = False,
            bbox_inches = 'tight');
```

```python
R = 1.5

ReG = [None] * 4
ImG = [None] * 4
L = [None] * 4
```

```python
# curve 1
L[0] = 8
ReG[0] = np.linspace(-R, R, L[0])[::-1]
ImG[0] = R * np.ones(L[0])

# curve 2
L[1] = 8
ReG[1] = -R * np.ones(L[1])
ImG[1] = np.linspace(-R, R, L[1])[::-1]

# curve 3
L[2] = 8
ReG[2] = np.linspace(-R, R, L[2])
ImG[2] = -R * np.ones(L[2])

# curve 4
L[3] = 8
ReG[3] = R * np.ones(L[2])
ImG[3] = np.linspace(-R, R, L[2])
```

```python
fig, ax = plt.subplots(figsize = (3.5, 3.5), dpi = 256)

ax.set_xlim(left = -2.5, right = 2.5)
ax.set_ylim(bottom = -2.5, top = 2.5)

for i in range(4):
    ax.quiver(ReG[i][:-1], ImG[i][:-1],
              np.diff(ReG[i]),
              np.diff(ImG[i]), color = 'C0')
    
for i in range(-2,3):
    ax.scatter(x = i, y = 0, color = 'C1')

fig.savefig('D4VSp47EOR.jpeg', transparent = False,
            bbox_inches = 'tight');
```
