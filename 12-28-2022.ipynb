```python
from sympy import *
```

\begin{equation*}
\int_{0}^{2\pi} \frac{1}{5+4 \sin \theta} {\rm d} \theta = \frac{2\pi}{3}
\end{equation*}

```python
z = Symbol('z', complex = True)
S = (z - 1/z) / (2 * I)
f = factor(1 / ((5 + 4 * S) * (I * z)) )
f
```

```python
2 * pi * I * residue(f, x = z, x0 = -I/2)
```

\begin{equation*}
\int_{-\pi}^{\pi}\frac{1}{1+\sin^2 \theta}{\rm d}\theta
\end{equation*}

```python
f = factor(1 / ((1 + S**2) * (I * z)) )
f
```

```python
P1 = 1 - sqrt(2)
P2 = -P1
R1 = residue(f, x = z, x0 = P1)
R2 = residue(f, x = z, x0 = P2)
simplify(2 * pi * I * (R1 + R2))
```
