```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 3.0; r = 1.0; h = 0.25

ReG = [None] * 4
ImG = [None] * 4
```

```python
# curve 1
a = np.arctan(h / np.sqrt(R**2 + h**2))
b = 2.0 * np.pi - a
t = np.linspace(a, b, 40)
ReG[0] = R * np.cos(t)
ImG[0] = R * np.sin(t)
```

```python
# curve 2
a = np.arctan(h / np.sqrt(r**2 + h**2))
b = 2.0 * np.pi - a
t = np.linspace(a, b, 13)[::-1]
ReG[1] = r * np.cos(t)
ImG[1] = r * np.sin(t)
```

```python
# curve 3
ReG[2] = np.linspace(ReG[1][-1], ReG[0][-1], 6)[::-1]
ImG[2] = -h * np.ones(6)
```

```python
# curve 4
ReG[3] = np.linspace(ReG[1][0], ReG[0][0], 6)
ImG[3] = h * np.ones(6)
```

```python
fig, ax = plt.subplots(figsize = (3.5, 3.5), dpi = 256)

ax.set_xlim(left = -3.5, right = 3.5)
ax.set_ylim(bottom = -3.5, top = 3.5)

for i in range(4):
    ax.quiver(ReG[i][:-1], ImG[i][:-1],
                np.diff(ReG[i]),
                np.diff(ImG[i]), color = 'C0')
```
